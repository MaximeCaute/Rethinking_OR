```python
def merc_from_arrays(lats, lons):
    r_major = 6378137.000
    x = r_major * np.radians(lons)
    scale = x/lons
    y = 180.0/np.pi * np.log(np.tan(np.pi/4.0 + lats * (np.pi/180.0)/2.0)) * scale
    return (x, y)
```

Representation of the nodes in a 20x20 matrix

```python
def generate_mat(df, size=0):
    lat = df['rel_lat'].apply(lambda x: int(x)).to_numpy()
    lon = df['rel_lon'].apply(lambda x: int(x)).to_numpy()

    if size == 0:
        size = max(max(lon), max(lat))
        
    nodes_mat = np.zeros((size+1,size+1))
    nodes_mat[lat, lon] = 1
    return nodes_mat

nodes_mat =generate_mat(df_nodes)
```

```python
def distance(s_lat, s_lng, e_lat, e_lng):
    """
    Compute distance with Haversine function
    """
    
    # approximate radius of earth in km
    R = 6373.0
    
    s_lat = s_lat*np.pi/180.0                      
    s_lng = np.deg2rad(s_lng)     
    e_lat = np.deg2rad(e_lat)                       
    e_lng = np.deg2rad(e_lng)  
    
    d = np.sin((e_lat - s_lat)/2)**2 + np.cos(s_lat)*np.cos(e_lat) * np.sin((e_lng - s_lng)/2)**2
    
    return 2 * R * np.arcsin(np.sqrt(d)) 
```

```python
distance1 = distance(min_lat, min_long, max_lat, min_long)
distance2 = distance(min_lat, min_long, min_lat, max_long)

# total area covered in km2
tot_distance = distance1*distance2


def resolution(height, width, min_lat, min_lon, max_lat, max_lon):
    lon = distance(min_lat,min_lon, min_lat, max_lon)*1000 # max width covered in m
    lat = distance(min_lat,min_lon, max_lat, min_lon)*1000 # max height covered in m
    r_h = lat/height
    r_w = lon/width
    
    pixel = r_h*r_w # area of the pixel in m2
    
    return r_h, r_w, pixel
```
