# Output visualization - Evolution of results during training

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(5, 3, figsize=(20,25))
#[0, 5, 10, 15, 20]
for idx, e in enumerate([0,5, 10, 20, 40]):
    a = np.load('./images_output/target_e_weight_99{}.npy'.format(e))
    b = np.load('./images_output/output_e_weight_99{}.npy'.format(e))
    c = np.load('./images_output/input_e_weight_99{}.npy'.format(e))
    ax[idx, 0].imshow(a)
    ax[idx, 0].set_title('Target. Epoch: {}'.format(e))
    ax[idx, 1].imshow(np.clip(c[1], 0, 1))
    ax[idx, 1].set_title('Input. State. Epoch: {}'.format(e))
    
    ax[idx, 2].imshow(np.argmax(b, axis=0))
    ax[idx, 2].set_title('Output. Argmax. Epoch: {}'.format(e))
    
plt.show()
```

```python
fig, ax = plt.subplots(5, 4, figsize=(20,25))
#[0, 5, 10, 15, 20]
for idx, e in enumerate([0, 20, 30, 40, 50]):
    inp = np.load('./images_output/input_e_weight{}.npy'.format(e))
    a = np.load('./images_output/target_e_weight{}.npy'.format(e))
    b = np.load('./images_output/output_e_weight{}.npy'.format(e))
    ax[idx, 0].imshow(a)
    ax[idx, 0].set_title('Target. Epoch: {}'.format(e))
    ax[idx, 1].imshow(b[1])
    ax[idx, 1].set_title('Output. Second channel. Epoch: {}'.format(e))
    ax[idx, 2].imshow(np.clip((inp[1] + inp[0]), 0, 1))
    ax[idx, 2].set_title('Input. Epoch: {}'.format(e))
    
    ax[idx, 3].imshow(np.argmax(b, axis=0))
    ax[idx, 3].set_title('Output. Argmax. Epoch: {}'.format(e))
    
plt.show()
```

```python
fig, ax = plt.subplots(5, 4, figsize=(20,25))
#[0, 5, 10, 15, 20]
for idx, e in enumerate([0, 20, 40, 60, 80]):
    inp = np.load('./images_output/input_e_weight_99{}.npy'.format(e))
    a = np.load('./images_output/target_e_weight_99{}.npy'.format(e))
    b = np.load('./images_output/output_e_weight_99{}.npy'.format(e))
    ax[idx, 0].imshow(a)
    ax[idx, 0].set_title('Target. Epoch: {}'.format(e))
    ax[idx, 1].imshow(b[1])
    ax[idx, 1].set_title('Output. Second channel. Epoch: {}'.format(e))
    ax[idx, 2].imshow(np.clip((inp[1]*(-1) + inp[0]), -1, 1))
    ax[idx, 2].set_title('Input. Epoch: {}'.format(e))
    
    ax[idx, 3].imshow(np.argmax(b, axis=0))
    ax[idx, 3].set_title('Output. Argmax. Epoch: {}'.format(e))
    
plt.show()
```
