# TRAINING NN

```python
import torch.nn as nn
import torch
import numpy as np
import pandas as pd
import random
import tqdm

import matplotlib.pyplot as plt

from torch.utils.data import Dataset
from torch import optim
```

```python
class ImagesP2(nn.Module):
    def __init__(self, inp):
        super().__init__()

        ins = 4
        cs = 5
        
        self.conv1 = nn.Sequential(
            nn.Conv2d(inp, ins, kernel_size=cs),
            nn.BatchNorm2d(ins),
            nn.ReLU(),
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(ins, ins, kernel_size=cs),
            nn.BatchNorm2d(ins),
            nn.ReLU(),
        )
        self.conv3 = nn.Sequential(
            nn.Conv2d(ins, ins, kernel_size=cs),
            nn.BatchNorm2d(ins),
            nn.ReLU(),
        )
        self.conv4 = nn.Sequential(
            nn.Conv2d(ins, ins, kernel_size=cs),
            nn.BatchNorm2d(ins),
            nn.ReLU(),
        ) 
        self.conv5 = nn.Sequential(
            nn.ConvTranspose2d(ins, ins, kernel_size=cs),
            nn.BatchNorm2d(ins),
            nn.ReLU(),
        )
        self.conv6 = nn.Sequential(
            nn.ConvTranspose2d(ins, ins, kernel_size=cs),
            nn.BatchNorm2d(ins),
            nn.ReLU(),
        )
        self.conv7 = nn.Sequential(
            nn.ConvTranspose2d(ins, ins, kernel_size=cs),
            nn.BatchNorm2d(ins),
            nn.ReLU(),
        )
        self.conv8 = nn.Sequential(
            nn.ConvTranspose2d(ins, 2, kernel_size=cs),
            nn.BatchNorm2d(2),
            nn.ReLU(),
        )

    def forward(self, x):
                
        
        x0 = x
        x1 = self.conv1(x0)
        x2 = self.conv2(x1)
        x3 = self.conv3(x2)
        x4 = self.conv4(x3)
        x5 = self.conv5(x4) + x3
        x6 = self.conv6(x5) + x2
        x7 = self.conv7(x6) + x1
        x8 = self.conv8(x7)
        x = x8
        return x
```

```python
MAX_V = 30
mini_batch_size = 50
#n_epochs=300
n_epochs=102

tables = list(range(1,21))
random.shuffle(tables)


train_tables, test_tables, validation_tables = tables[:15], tables[16:19], tables[19:]
```

```python
lr = 0.001

model = ImagesP2(31)
optimizer = optim.Adam(model.parameters(), lr=lr)
criterion = nn.CrossEntropyLoss(weight=torch.tensor([0.01, 0.99]))
#criterion = nn.CrossEntropyLoss()
#criterion = nn.MSELoss()
```

```python
def epoch_train(model, train_table, test_tables, optimizer, criterion, e, X, Y):

    sum_loss = 0
    distances_list = []
    acc = []

    # train data
    for k, TABLE in enumerate(train_tables):
        #print(TABLE)
        print('\rCompleted {:.2f}%'.format((k+1)/len(train_tables)*100), end="")
        data = np.load('./simulation/data_compact_{}.npy'.format(TABLE), allow_pickle=True).tolist()
        #info = pd.read_pickle('./simulation/comparison_{}.pkl'.format(TABLE))

        #x_shape = np.array(data[0][0]).shape

        idx = list(range(1,len(data)))
        random.shuffle(idx)
        
        


        for b in range(0, len(idx), mini_batch_size):

            t_idx = idx[b:b+mini_batch_size]
            x = np.asarray([data[i] for i in t_idx])
            x[:, 2, :, :] = np.clip(x[:, 2, :, :], 0 , 1)
            train_x = torch.tensor(x[:,:2, :,:]).type(torch.FloatTensor)
            train_y = torch.tensor(x[:,2,:,:]).type(torch.LongTensor)
            #train_y = torch.tensor(x[:,2,:,:]).type(torch.FloatTensor)
            
            #compute distance matrix
            _, row, col = np.where(train_y == 1)
            
            X_copy = X.repeat(train_y.shape[0], 1, 1)
            Y_copy = Y.repeat(train_y.shape[0], 1, 1)
            
            row_X = torch.tensor(row.repeat(train_y.shape[1]**2)).reshape(train_y.shape[0],train_y.shape[1], train_y.shape[1])
            col_Y = torch.tensor(col.repeat(train_y.shape[1]**2)).reshape(train_y.shape[0],train_y.shape[1], train_y.shape[1])
            
            dist = (X_copy - row_X)**2 + (Y_copy - col_Y)**2
            
            # set gradient to zero
            optimizer.zero_grad()
            
            # compute output
            output = model(train_x)
            batch_loss = criterion(output, train_y)
            
            batch_loss.backward()
            optimizer.step()

            sum_loss = sum_loss + batch_loss.item()
            
            acc.append(100*output.detach().max(1)[1].eq(train_y).sum().item()/(train_y.shape[0]*train_y.shape[1]*train_y.shape[2]))
            
            # compute vehicle assignment
            #dist_out = dist.mul(torch.tensor(np.argmax(output.detach().numpy(), axis=1)))
            #idx_pos, row_pos, col_pos = np.where(dist_out>0)
            
            #for i in range(train_y.shape[0]):
            #    r = row_pos[idx_pos == i]
            #    c = col_pos[idx_pos == i]
                
                # any vehicle is assigned
                #if (len(r) == 0):
                    #min_dist = train_y.shape[1]**2
                    #v_r, v_c = 0, 0
                #else:   
                    #min_dist = min(dist_out[i,r,c])
                    #min_val = np.argmin(dist_out[i, r, c])
                    #v_r, v_c = r[min_val], c[min_val]

                #distances_list.append(min_dist)
                
                #real_v_r, real_v_c = np.where(train_y[i] == 1)
                #if v_r == real_v_r and v_c == real_v_c:
                    #acc += 1
                
                
            
            
            
            
    
    print('\nEpoch {}. Train Loss: {:.3f}'.format(e+1,sum_loss))
    #print('Epoch {}. Mean distance: {:.3f}'.format(e+1,np.mean(distances_list)))
    print('Epoch {}. Accuracy: {:.3f}'.format(e+1,np.mean(acc)/len(acc)))
    
    
    test_loss = 0
    
    for k,TABLE in enumerate(test_tables):
        data = np.load('./simulation/data_compact_{}.npy'.format(TABLE), allow_pickle=True).tolist()
        idx = list(range(1,len(data)))
        random.shuffle(idx)
        
        for b in range(0, len(idx), mini_batch_size):

            t_idx = idx[b:b+mini_batch_size]
            x = np.asarray([data[i] for i in t_idx])
            x[:, 2, :, :] = np.clip(x[:, 2, :, :], 0 , 1)
            
            test_x = torch.tensor(x[:,:2, :,:]).type(torch.FloatTensor)
            test_y = torch.tensor(x[:,2,:,:]).type(torch.LongTensor)
            #test_y = torch.tensor(x[:,2,:,:]).type(torch.FloatTensor)
            
            output = model(test_x)
            batch_loss = criterion(output, test_y)
            

            test_loss += batch_loss.item()
    if e%5 == 0:
        inp = test_x[-1]
        a = test_y[-1]
        b = output[-1].detach().numpy()
        
        np.save('./images_output/input_e_weight_99{}'.format(e), inp.numpy())
        np.save('./images_output/target_e_weight_99{}'.format(e), a.numpy())
        np.save('./images_output/output_e_weight_99{}'.format(e), b)

    print('Epoch {}. Test Loss: {:.3f}'.format(e+1,test_loss))
    
    return sum_loss, np.mean(acc)/len(acc)
```
